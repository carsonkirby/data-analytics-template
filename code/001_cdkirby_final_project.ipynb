```python
import pandas as pd
```

#### The county_data.csv file was the file that was created and downloaded after using web scraping on the Iowa State University website. 

```python
population_change = pd.read_csv('county_data.csv')
```

```python
display(population_change)
```

```python
# Renaming 'Area Name' to 'County' to prepare for merging. I also dropped the column titled FIPS, because it was not needed for the purposes of this project
population_change = population_change.rename(columns={'Area Name': 'County', '1980':'1980_pop', '1990':'1990_pop', '2000':'2000_pop', '2010':'2010_pop', '2020':'2020_pop'})
population_change = population_change.drop(columns=['FIPS'])
population_change
```

```python
# The income.csv file was downloaded from the National Institute on Minority Health and Health Disparites website
income = pd.read_csv('income.csv')
income = income.rename(columns={'Value (Dollars)':'Median_Household_Income'})
```

```python
display(income)
```

```python
# The unemployment.csv file was also downloaded from the National Institute on Minority Health and Health Disparites website. I removed the unnecessary columns. 
unemployment = pd.read_csv('unemployment.csv')
unemployment = unemployment.drop(columns=['FIPS'])
unemployment = unemployment.drop(columns=['Rank within US (of 3143 counties)'])
unemployment = unemployment.drop(columns=['People Unemployed)'])
```

```python
display(unemployment)
```

```python
rural_urban = pd.read_csv('rural_urban.csv')
```

```python
display(rural_urban)
```

```python
integrated_counties = pd.merge(population_change, income, on="County", how="inner")
```

```python
display(integrated_counties)
```

```python
integrated_counties1 = pd.merge(integrated_counties, unemployment, on="County", how="inner")
```

```python
integrated_counties1
```

```python
integrated_counties2 = pd.merge(integrated_counties1, rural_urban, on="County", how="inner")
integrated_counties2['County'] = integrated_counties2['County'].str.replace(r' County$', '', regex=True)
```

```python
integrated_counties2
```

```python
integrated_counties2.to_csv("integrated_counties.csv", header=True, index=False, encoding="utf-8")
```
